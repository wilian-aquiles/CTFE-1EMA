#**Exemplo- Cálculo de variáção e desvio padão **

```python
n1, n2, n3, n4 = input("Insira os números da distribuição: ").split()
n1 = int(n1)
n2 = int(n2)
n3 = int(n3)
n4 = int(n4)
```

```python
#Média
media = ( n1 + n2 + n3 + n4 )/4
media
```

```python
#Desvio

d1 = n1 - media
d2 = n2 - media
d3 = n3 - media
d4 = n4 - media

d1,d2,d3,d4
```

```python
#Somatória do quadrado dos desvios
soma = d1**2 + d2**2 + d3**2 + d4**2
#Variância da distribuição
variancia = soma / 4

variancia
```

```python
#Desvio padrão
DP = variancia**0.5

DP
```
